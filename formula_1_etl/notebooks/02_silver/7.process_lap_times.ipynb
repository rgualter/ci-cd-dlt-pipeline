### Process Lap Times

```python
from formula1.formula1_constants import *
from formula1.formula1_utils import *
```

##### Step 1 - Read from Bronze Layer

```python
lap_times_df = spark.read.table("f1_bronze.lap_times")
```

##### Step 2 - Rename Columns

```python
from pyspark.sql.functions import col

lap_times_renamed_df = lap_times_df.withColumnRenamed(
    "driverId", "driver_id"
).withColumnRenamed("raceId", "race_id")
```

##### Step 3 - Add Ingestion Date

```python
lap_times_final_df = add_ingestion_date(lap_times_renamed_df)
```

##### Step 4 - Write to Silver Layer

```python
lap_times_final_df.write.mode("overwrite").partitionBy("race_id").format(
    "delta"
).saveAsTable("f1_silver.lap_times")
```
