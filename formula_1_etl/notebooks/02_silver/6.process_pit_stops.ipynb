### Process Pit Stops

```python
from formula1.formula1_constants import *
from formula1.formula1_utils import *
```

##### Step 1 - Read from Bronze Layer

```python
pit_stops_df = spark.read.table("f1_bronze.pit_stops")
```

##### Step 2 - Rename Columns

```python
from pyspark.sql.functions import col

pit_stops_renamed_df = pit_stops_df.withColumnRenamed(
    "driverId", "driver_id"
).withColumnRenamed("raceId", "race_id")
```

##### Step 3 - Add Ingestion Date

```python
pit_stops_final_df = add_ingestion_date(pit_stops_renamed_df)
```

##### Step 4 - Write to Silver Layer

```python
pit_stops_final_df.write.mode("overwrite").partitionBy("race_id").format(
    "delta"
).saveAsTable("f1_silver.pit_stops")
```
