### Process Qualifying

```
from src.formula1.formula1_constants import *
from src.formula1.formula1_utils import *
```

##### Step 1 - Read from Bronze Layer

```
qualifying_df = spark.read.table("f1_bronze.qualifying")
```

##### Step 2 - Rename Columns

```
from pyspark.sql.functions import col

qualifying_renamed_df = (
    qualifying_df.withColumnRenamed("qualifyId", "qualify_id")
    .withColumnRenamed("driverId", "driver_id")
    .withColumnRenamed("raceId", "race_id")
    .withColumnRenamed("constructorId", "constructor_id")
)
```

##### Step 3 - Add Ingestion Date

```
qualifying_final_df = add_ingestion_date(qualifying_renamed_df)
```

##### Step 4 - Write to Silver Layer

```
qualifying_final_df.write.mode("overwrite").partitionBy("race_id").format(
    "delta"
).saveAsTable("f1_silver.qualifying")
```
