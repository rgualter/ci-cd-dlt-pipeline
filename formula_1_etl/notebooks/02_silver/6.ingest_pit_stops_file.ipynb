### Silver Layer: Process pit_stops data

Read from bronze/raw layer (parquet) and transform for silver/processed layer.

```python
dbutils.widgets.text("p_data_source", "")
v_data_source = dbutils.widgets.get("p_data_source")
```

```python
from formula1.formula1_constants import raw_folder_path, processed_folder_path
```

```python
from formula1.formula1_utils import add_ingestion_date
```

##### Step 1 - Read parquet from bronze/raw layer

```python
pit_stops_df = spark.read.parquet(f"{raw_folder_path}/pit_stops")
```

##### Step 2 - Rename columns and add new columns
1. Rename driverId and raceId
1. Update ingestion_date

```python
pit_stops_with_ingestion_date_df = add_ingestion_date(pit_stops_df)
```

```python
final_df = pit_stops_with_ingestion_date_df \
    .withColumnRenamed("driverId", "driver_id") \
    .withColumnRenamed("raceId", "race_id")
```

##### Step 3 - Write to output to silver/processed layer in parquet format

```python
final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/pit_stops")
```
