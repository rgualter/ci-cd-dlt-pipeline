### Process Drivers

```
from src.formula1.formula1_constants import *
from src.formula1.formula1_utils import *
```

##### Step 1 - Read from Bronze Layer

```
drivers_df = spark.read.table("f1_bronze.drivers")
```

##### Step 2 - Transform and Rename
- Concat name
- Rename IDs

```
from pyspark.sql.functions import col, concat, lit

drivers_transformed_df = (
    drivers_df.withColumnRenamed("driverId", "driver_id")
    .withColumnRenamed("driverRef", "driver_ref")
    .withColumn("name", concat(col("name.forename"), lit(" "), col("name.surname")))
)

drivers_selected_df = drivers_transformed_df.select(
    "driver_id", "driver_ref", "number", "code", "name", "dob", "nationality"
)
```

##### Step 3 - Add Ingestion Date

```
drivers_final_df = add_ingestion_date(drivers_selected_df)
```

##### Step 4 - Write to Silver Layer

```
drivers_final_df.write.mode("overwrite").format("delta").saveAsTable(
    "f1_silver.drivers"
)
```
