### Silver Layer: Process constructors data

Read from bronze/raw layer (parquet) and transform for silver/processed layer.

```python
dbutils.widgets.text("p_data_source", "")
v_data_source = dbutils.widgets.get("p_data_source")
```

```python
from formula1.formula1_constants import raw_folder_path, processed_folder_path
```

```python
from formula1.formula1_utils import add_ingestion_date
```

##### Step 1 - Read parquet from bronze/raw layer

```python
constructor_df = spark.read.parquet(f"{raw_folder_path}/constructors")
```

##### Step 2 - Drop unwanted columns from the dataframe

```python
from pyspark.sql.functions import col
```

```python
constructor_dropped_df = constructor_df.drop(col('url'))
```

##### Step 3 - Rename columns and update ingestion date

```python
constructor_renamed_df = constructor_dropped_df \
    .withColumnRenamed("constructorId", "constructor_id") \
    .withColumnRenamed("constructorRef", "constructor_ref")
```

```python
constructor_final_df = add_ingestion_date(constructor_renamed_df)
```

##### Step 4 - Write output to silver/processed layer in parquet format

```python
constructor_final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/constructors")
```
