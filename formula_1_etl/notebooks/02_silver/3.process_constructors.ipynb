### Process Constructors

```python
from formula1.formula1_constants import *
from formula1.formula1_utils import *
```

##### Step 1 - Read from Bronze Layer

```python
constructors_df = spark.read.table("f1_bronze.constructors")
```

##### Step 2 - Select and Rename Columns

```python
from pyspark.sql.functions import col

# Drops url by not selecting it
constructors_selected_df = constructors_df.select(
    col("constructorId").alias("constructor_id"),
    col("constructorRef").alias("constructor_ref"),
    col("name"),
    col("nationality"),
)
```

##### Step 3 - Add Ingestion Date

```python
constructors_final_df = add_ingestion_date(constructors_selected_df)
```

##### Step 4 - Write to Silver Layer

```python
constructors_final_df.write.mode("overwrite").format("delta").saveAsTable(
    "f1_silver.constructors"
)
```
