# Bronze Layer: Execute all ingestion notebooks

This notebook orchestrates the ingestion of raw files (CSV/JSON) from the landing zone into the bronze/raw layer as parquet files.

```python
dbutils.widgets.text("p_data_source", "")
v_data_source = dbutils.widgets.get("p_data_source")
```

```python
notebooks = [
    "1.bronze_circuits",
    "2.bronze_races",
    "3.bronze_constructors",
    "4.bronze_drivers",
    "5.bronze_results",
    "6.bronze_pit_stops",
    "7.bronze_lap_times",
    "8.bronze_qualifying"
]

for notebook in notebooks:
    print(f"Executing {notebook}...")
    result = dbutils.notebook.run(notebook, 0, {"p_data_source": v_data_source})
    print(f"Result: {result}")
```
